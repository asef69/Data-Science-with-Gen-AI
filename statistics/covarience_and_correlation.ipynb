```python
#covarience and correlation always will be calculated on numerical data only. Hence we will select only numerical data from the dataset.

numerical_df=df.select_dtypes(include=['int64','float64'])
```

```python
numerical_df
```

```python
numerical_df.cov()
```

```python
numerical_df.corr()
```

```python
df.corr(numeric_only=True)
```

```python
#plot heatmap for correlation
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
#plot heatmap for covarience
sns.heatmap(numerical_df.cov(),annot=True)
```

```python
#plot graph for correlation between total_bill and tip
sns.scatterplot(x=df['total_bill'],y=df['tip'])
```
