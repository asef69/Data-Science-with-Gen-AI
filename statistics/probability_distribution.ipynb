```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#normal distribution>>gaussian distribution>>majority of data is clustered towards the mean

mu=0
sigma=0.1

s=np.random.normal(mu,sigma,size=10000)

```

```python
len(s)
```

```python
plt.hist(s,30)
plt.show()
```

```python
sns.distplot(s)
plt.show()
```

```python
#uniform distribution>> all the outcomes has equal probability of occuring in a range
u = np.random.uniform(0,1,10000)
plt.hist(u)
plt.show()
```

```python
sns.distplot(u)
plt.show()
```

```python
#log normal distribution
ln = np.random.lognormal(mean=0,sigma=0.5,size=10000)
ln
```

```python
ln=ln[np.where(abs(ln)<10)]
plt.hist(ln,bins=100)
plt.show()
```

```python
sns.boxplot(ln)
#since log normal distribution is right skewed, it has more outliers on the right side of the boxplot
```

```python
sns.displot(ln)
```

```python
#lognormal distribution becomes normal distribution when we take log of the data
sns.distplot(np.log(ln))
```

```python
#square root transformation also helps in reducing the skewness of the data
sns.distplot(np.sqrt(ln))
```

```python
#chi square distribution
chi = np.random.chisquare(df=30,size=10000)
chi
```

```python
chi=chi[np.where(abs(chi)<100)]
```

```python
plt.hist(ln,bins=1000)
plt.show()
```

```python
#multiple chi square distribution into one frame
d=np.random.chisquare(df=10,size=10000)
d=d[np.where(abs(d)<100)]
plt.hist(d,bins=1000)
plt.hist(chi,bins=1000)
plt.show()
```

```python
# t distribution
t = np.random.standard_t(df=3,size=10000)
t=t[np.where(abs(t)<100)]
plt.hist(t,bins=1000)
plt.show()
```

```python
t = np.random.standard_t(df=20,size=10000)
t=t[np.where(abs(t)<100)]
plt.hist(t,bins=1000)
plt.show()
```

```python
#f distribution
f = np.random.f(dfnum=100,dfden=100,size=10000)
f=f[np.where(abs(f)<100)]
plt.hist(f,bins=10000)
plt.show()
```

```python
#discrete distribution
#bernoulli distribution>>binomial distribution with one trail is called barnoulli distribution

b=np.random.binomial(n=1,p=0.5,size=10000)
plt.hist(b)
plt.show()
```

```python
#binomial distribution>> number of success in a sequence of n independent yes/no experiment
n=10
p=0.5
binomial=np.random.binomial(n,p,size=10000)
plt.hist(binomial)
plt.show()
```

```python
#poisson distribution
p=np.random.poisson(lam=5,size=10000)
plt.hist(p,bins=20) 
plt.show()
```
