```python
#mean
age=[25,28,37,24,32]
mean=sum(age)/len(age)
print("Mean:",mean)
```

```python
import numpy as np
np.mean(age)
```

```python
height=[182.5,175.2,168.4,190.3,185.6]
mean_height=np.mean(height)
print("Mean Height:",mean_height)
```

```python
#median
nos=[2,3,5,6,8,3,5,6,9,10]
median=np.median(nos)
print("Median:",median)
```

```python
from scipy import stats
mode=stats.mode(nos)
print("Mode using scipy:",mode)
```

```python
import statistics
mode_stat=statistics.mode(nos)
print("Mode using scipy:",mode_stat)
```

```python
height=[5.5,6.0,5.8,5.9,6.1,12]
mean_height=np.mean(height)
print("Mean Height:",mean_height)
#mean is affected by outliers, it is not a good measure of central tendency in this case. We can use median instead.
median_height=np.median(height)
print("Median Height:",median_height)
```
