```python
data=[2,3,5,65,54,65,22,58,23,67]
```

```python
#measure of dispersion/spread
#1.range
max(data)-min(data)
```

```python
#2.percentiles
import numpy as np
np.percentile(data,[0,25,50,75,100])
```

```python
#3.IQR
IQR=np.percentile(data,75)-np.percentile(data,25)
IQR
```

```python
#4.boxplot
import seaborn as sns
sns.boxplot(data) #upper fence>> Q3+1.5*IQR
#lower fence>> Q1-1.5*IQR
```

```python
#5.variance
np.var(data)
```

```python
#6.standard deviation
np.std(data)
```

```python
import statistics
statistics.variance(data)
#sample variance>> denominator n-1
```

```python
statistics.stdev(data)
```

```python
statistics.pvariance(data)
```

```python
#measure of symmericity
#skewness
from scipy.stats import skew
skew(data)
```

```python
#value is between -1 to +1>> symmetric distribution
#value is >+1>> positively skewed distribution
#value is <-1>> negatively skewed distribution
#plot the distribution to confirm the skewness
sns.histplot(data,kde=True)
```
