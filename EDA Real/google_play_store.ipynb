```python
#problem is we have various types of size in size column, we need to convert it into one type of size. Also we have a data called "varies with device" which is not a size, we need to remove that data from size column. We will convert all the sizes into MB(Megabyte) unit. We have sizes in KB(Kilobyte) and GB(Gigabyte) unit, we will convert them into MB unit. 1GB=1024MB and 1KB=0.001MB
#now we will write a function to do the whole work 
#1MB=1024KB
#1GB=1024MB
def size_process(size):
    if size.endswith('M'):
        return float(size[:-1])*1024
    elif size.endswith('k'):
        return float(size[:-1])
    elif size.endswith('G'):
        return float(size[:-1])*1024*1024
    else:
        return str(np.nan)
```

```python
item="19M"
print(size_process(item))
item="14k"
print(size_process(item))
item="1.5G"
print(size_process(item))
item="Varies with device"
print(size_process(item))
```

```python
df_copy.Size.apply(size_process) #applying the function to size column
```
