```python
#property decorators >> it is used to access the private attributes using getter method without calling it as a method.
class Student:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age
    @property
    def get_name(self): #access private attribute
        return self.__name   

    @get_name.setter 
    def set_name(self,name): #set private attribute
        self.__name=name

    @get_name.deleter
    def del_name(self):
        del self.__name    
```

```python
class Circle:
    def __init__(self,radius):
        self.__radius=radius

    @property    
    def radius(self):
        return self.__radius

    @radius.setter
    def radius(self,value):
        if value<=0:
                raise ValueError("Radius is not positive")

        else: self.__radius=value 

    def area(self):
        return 3.14*self.__radius*self.__radius           
```

```python
c=Circle(5)
```

```python
print(c.radius)
```

```python
c.radius=10
```

```python
c.area()
```
