```python
#__init__.py is a dunder method used to initialize the attributes of a class when an object is created.
class Student:
    def __new__(cls, name, age):
        print("Creating a new Student object")
        return super(Student, cls).__new__(cls)
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
obj=Student("John",20)
```
