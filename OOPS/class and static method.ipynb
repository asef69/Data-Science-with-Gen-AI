```python
class Student:
    #constructor
    def __init__(self, name):#creating instance
        self.name = name
    @classmethod
    def student_details (cls,name):#overloading the init method.even if we change the parameter name, it will still work 
        return cls(name)
```

```python
obj1 = Student.student_details("Alice")
```

```python
obj1.name
#class method is alternatively used to create instances of a class using different parameters. with out __init__ method. 
```

```python
#class method can modify class state applicable to all the instances of the class 
class Student:
    total_student=0
    def __init__(self,name):
        self.name=name
        Student.total_student +=1

    @classmethod
    def get_total_students(cls):
        return cls.total_student
```

```python
Student.total_student
```

```python
std1=Student("John")
std2=Student("Alice")
Student.get_total_students()
```

```python
#static method can be used to create utility functions.
class Student:
    total_student=0
    def __init__(self,name):
        self.name=name
        Student.total_student +=1

    @classmethod
    def get_total_students(cls):
        return cls.total_student

    @staticmethod
    def is_adult(age):
        return age>=18
```

```python
Student.student_details("Alice") #already deleted
```

```python
Student.is_adult(20)
```
