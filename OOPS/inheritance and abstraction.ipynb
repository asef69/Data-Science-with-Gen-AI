```python
class Parent:
    def greet(self):
        return "Hello from Parent"
```

```python
class Child(Parent):
    def greet(self):
        parent_greeting = super().greet()
        return f"{parent_greeting} and Hello from Child"
```

```python
child_object=Child()
print(child_object.greet())
```

```python
class Fruit:
    def fruit_info(self):
        print("Inside Parent Fruit class")

class Apple(Fruit):
    def fruit_info(self):
        print("Inside Child Apple class")

    def apple_variety(self):
        print("This is a variety of Apple")            
```

```python
obj=Apple()
obj.fruit_info()
obj.apple_variety()
```

```python
import abc
```

```python
class PWskills(abc.ABC):
    @abc.abstractmethod
    def course_info(self):
        print("This method provides course information")
```

```python
PWskills().course_info()
```
