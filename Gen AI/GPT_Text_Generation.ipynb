```python
#Example-3: Multi turn Conversation
conversations=[]
queries=[
    "I'm planning a trip to Japan.What are the top 3 cities I should visit?",
    "For the first city,what are some must see attractions?",
    "What's the best time of the year to visit the city?",
    "Can you recomment some local dishes to try there?",
    "How do I get from this city to the second city you recommend?"
]
for query in queries:
  conversations.append({"role":"user","content":query})
  response=client.chat.completions.create(
    model='gpt-4o',
    messages=conversations
  )
  assistant_response=response.chpices[0].message.content
  conversations.append({"role":"assistant","content":assistant_response})
  display_markdown(f"Human Query: {query}")
  display_markdown(f"Assistant Response: {assistant_response}")
```
