```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

```

```python
model_name="gpt2"
tokenizer=GPT2Tokenizer.from_pretrained(model_name)
model=GPT2LMHeadModel.from_pretrained(model_name)
```

```python
def generate_poem(prompt,max_length=100):
  #Encoding the prompt using our tokenizer
  input_ids=tokenizer.encode(prompt,return_tensors="pt")

  output=model.generate(
      input_ids,
      max_length=max_length,
      num_return_sequences=1,
      no_repeat_ngram_size=2,
      top_k=50,
      top_p=0.95,
      temperature=0.7
  )

  generated_text=tokenizer.decode(output[0],skip_special_token=True)

  return generated_text
```

```python
prompt="Twinkle twinkle little star"
poem=generate_poem(prompt)
print("----------------POEM-----------------")
print(poem)
```

#Part -2 :Music Generation

```python

from music21 import stream, note, scale
import random

def generate_melody(length=8, scake=scale.MajorScale('C')):
  melody = stream.Stream()
  for _ in range(length):

    pitch = random.choice(scake.getPitches("C4", "C5"))
    duration = random.choice([0.5, 1, 1.5, 2])
    new_note = note.Note(pitch)
    new_note.duration.quarterLength = duration
    melody.append(new_note)
  return melody
```

```python
melody=generate_melody()
melody.show('midi')
```

#Part 3: Combining Poetry and Music

Translate poetry to music

```python
def poem_to_melody(poem,scale=scale.MajorScale('A')):
  #Tokenize the poem
  words=poem.split()
  melody=stream.Stream()
  for work in words[:8]:
    pitch=scale.pitchFromDegree((len(work)%7)+1)
    duration=len(work)*0.25
    new_note=note.Note(pitch)
    new_note.duration.quarterLength=duration
    melody.append(new_note)
  return melody

```

```python
poem=generate_poem("The wind in whispers")
print("----------------POEM-----------------")
print(poem)

print("*************************************")
melody=poem_to_melody(poem)
melody.show('midi')
```
