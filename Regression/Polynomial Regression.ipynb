```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(1)
```

```python
X=2*np.random.rand(100,1)
```

```python
X
```

```python
y=4+3*X+1.5*X**2+np.random.rand(100,1)
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#perform polynomial 
from sklearn.preprocessing import PolynomialFeatures
regressor=PolynomialFeatures(degree=2,include_bias=False)
```

```python
regressor
```

```python
X_poly_train=regressor.fit_transform(X_train)
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
poly_regressor=LinearRegression()
```

```python
poly_regressor
```

```python
poly_regressor.fit(X_poly_train,y_train)
```

```python
poly_regressor.coef_
```

```python
poly_regressor.intercept_
```

```python
y_poly_pred=poly_regressor.predict(X_poly_train)
```

```python
y_poly_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train,y_poly_pred)
```

```python
plt.scatter(X_train,y_train,label='train')
```

```python
degree = 2
plt.scatter(X_train, y_train, label = "training data")
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = regressor.transform(X_range)
plt.plot(X_range, poly_regressor.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
