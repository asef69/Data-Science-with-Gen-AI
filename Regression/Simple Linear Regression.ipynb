```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the data
#here we are loading the data from scikit learn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

Problem Statement:
To predict disease progression one year after baseline provided 10 features

```python
diabetes.DESCR
```

```python
print(diabetes.DESCR)
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
data=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
```

```python
data['target']=diabetes.target
```

```python
data
```

```python
X=data[['bmi']]
y=data['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.,random_state=1)
```

```python
X_train
```

```python
#scaling
```

```python
#model training
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
#training a model
model.fit(X_train,y_train)
#at the backend it will use gradient descent to find optimal coefficient
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
#visualize the result
plt.scatter(X_test,y_test,color='black',label='Actual Data')
```

```python
plt.scatter(X_test,y_test,color='black',label='Actual Data')
plt.plot(X_test,y_pred,color='blue',linewidth=3,label='Predicted Linear Regression Line')
plt.xlabel('BMI')
plt.ylabel('One year Progression Target')
plt.title('Linear Regression on Diabetes Data')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
