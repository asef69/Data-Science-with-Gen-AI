```python
#imputation of missing values
df.isnull().sum()
```

```python
#age is missing at random
sns.histplot(df['age'], kde=True)
```

```python
#whenever we have normal distribution data and no outliers , we can impute with mean
df['mean_imputation']=df['age'].fillna(df['age'].mean())
```

```python
df[['age','mean_imputation']]
```

```python
sns.distplot(df['mean_imputation'], kde=True)
```

```python
#mean imputation works with normally distributed data and no outliers. if we have outliers then mean imputation will not work. in that case we can use median imputation
df['median_imputation']=df['age'].fillna(df['age'].median())
df[['age','median_imputation','mean_imputation']]
```

```python
sns.displot(df['median_imputation'], kde=True)
```
