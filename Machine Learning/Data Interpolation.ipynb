```python
#Data Interpolation is a process of estimatting unkonwn values from the known values
```

```python
#Linear Interpolation
#Cubic Interpolation
#Polynomial Interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9])

plt.scatter(x, y)
```

```python
#linear interplotaion>>linear in nature
x_new = np.linspace(1, 5, 10) #create new x values
y_interp = np.interp(x_new, x, y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp, color='red')
```

```python
#cubic interpolation>>non linear in nature
x=np.array([1, 2, 3, 4, 5])
y=np.array([1, 8, 27, 64, 125])

plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
f=interp1d(x, y, kind='cubic')
```

```python
f
```

```python
x_new=np.linspace(1, 5, 10)
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp, color='green')
```

```python
#polynomial interpolation>>non linear in nature
x=np.array([1, 2, 3, 4, 5])
y=np.array([1, 4, 9, 1, 25])

p=np.polyfit(x, y, deg=2)
```

```python
x_new=np.linspace(1, 5, 10)
y_interp=np.polyval(p, x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp, color='red')
```
