```python
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure, output_file, show

```

```python
flowers
```

```python
output_file("test.html")
p=figure(title="test_flower")
p.xaxis.axis_label='petal_length'
p.yaxis.axis_label='petal_width'
p.circle(flowers.petal_length, flowers.petal_width, size=10, color='navy', alpha=0.5)
show(p)
```

```python
#line chart
x=[1,2,3,4]
y=[10,15,13,17]
p=figure(title="Simple Line Chart", x_axis_label='Categories', y_axis_label='Values')
p.line(x, y, legend_label='Line', line_width=2)
show(p)
```
