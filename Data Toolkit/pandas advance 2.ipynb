```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv",header=None)
```

```python
df
```

```python
import numpy as np
df.groupby(['Pclass','Sex'])['Survived'].sum()
```

```python
# Multiple statistics
df.groupby(['Pclass', 'Sex'])['Survived'].agg(['sum', 'mean', 'count'])

# With custom names
df.groupby(['Pclass', 'Sex'])['Survived'].agg(
    total_survived='sum',
    survival_rate='mean',
    total_passengers='count'
)
```
