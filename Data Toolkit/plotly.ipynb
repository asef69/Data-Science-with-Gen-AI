```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%pip install plotly

import warnings
warnings.filterwarnings("ignore")
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4], y=[10,11,12,13], mode='lines+markers', name='lines+markers'))
fig.show()
```

```python
tip=sns.load_dataset("tips")
fig=go.Figure()
fig.add_trace(go.Scatter(x=tip.total_bill, y=tip.tip, mode='markers', name='markers'))
```

```python
#bar chart
fig=go.Figure()
fig.add_trace(go.Bar(x=tip.day, y=tip.total_bill, name='total_bill'))
fig.show()
```

```python
#histogram
fig=go.Figure()
fig.add_trace(go.Histogram(x=tip.total_bill, nbinsx=20, name='total_bill'))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4], y=[10,11,12,13], z=[5,6,7,8], mode='lines+markers', name='lines+markers'))
fig.show()
```

```python
x=['A','B','C','D']
y=[10,15,13,17]
fig=px.bar(x=x, y=y, title="Simple Bar Chart", labels={'x':'Categories', 'y':'Values'})
fig.show()
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
px.imshow(data, title="Heatmap Example", labels={'x':'X-axis', 'y':'Y-axis', 'color':'Intensity'})
```

```python
#bubble chart
import plotly.express as px
df = px.data.gapminder().query("year==2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)

fig.show()
```

```python
#3d scatter plot
fig = px.scatter_3d(df, x='gdpPercap', y='lifeExp', z='pop',
                    color='continent', size='pop', hover_name='country',
                    log_x=True, size_max=60)
fig.show()
```

```python
#density contour plot
fig = px.density_contour(tip, x="total_bill", y="tip",
                        title="Density Contour Plot of Total Bill vs Tip")
fig.show()
```
