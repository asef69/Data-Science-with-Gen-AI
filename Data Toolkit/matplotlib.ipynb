```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]
plt.plot(x,y1,color='blue',marker='o',linestyle='--',label='Prime Numbers')
plt.plot(x,y2,color='red',marker='s',linestyle='-.',label='Even Numbers')
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y=np.linspace(0,10,100)
plt.plot(y,np.sin(y),label='Sine Wave')
plt.plot(y,np.cos(y),label='Cosine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(y, np.sin(y), color='blue', label='Sine Wave')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('sin(X)')
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(y, np.cos(y), color='orange', label='Cosine Wave')
plt.title('Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('cos(X)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(y, np.sin(y), color='blue', label='Sine Wave')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('sin(X)')
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 2)
plt.plot(y, np.cos(y), color='orange', label='Cosine Wave')
plt.grid(True)
plt.title('Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('cos(X)')
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(y, np.tan(y), color='green', label='Tangent Wave')
plt.title('Tangent Wave')
plt.xlabel('X-axis')
plt.ylabel('tan(X)')
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 4)
plt.plot(y, np.sinh(y), color='red', label='Hyperbolic Sine Wave')
plt.title('Hyperbolic Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('sinh(X)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
a=np.random.rand(1000)
plt.hist(a, bins=48, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Numbers")
plt.show()
```

```python
#stacked histogram
a=np.random.rand(1000)
b=np.random.rand(1000)
plt.hist([a,b], bins=30, color=['skyblue','salmon'], edgecolor='black', stacked=True)
plt.title("Stacked Histogram of Two Random Number Sets")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x, y, color='magenta', marker='x', s=100, alpha=0.7)
plt.title("Scatter Plot of Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
df=pd.read_csv("bank_churn.csv")
```

```python
df
```

```python
plt.scatter(x=df['Age'], y=df['Balance'], c='blue', alpha=0.5)
```

```python
#3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Balance'], c='blue', alpha=0.5)
plt.show()
```

```python
#histogram
plt.hist(df['Age'], bins=30, color='lightgreen', edgecolor='black')
plt.title("Histogram of Customer Ages")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
#Box plot
plt.boxplot(df['Balance'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Box Plot of Customer Balances")
plt.show()
```

```python
#violin plot
plt.violinplot(df['Balance'], vert=True, showmeans=True, showmedians=True
)
plt.title("Violin Plot of Customer Balances")
plt.show()
```

```python
#area plot
plt.figure(figsize=(10, 6))
plt.fill_between(df.index, df['Balance'], color='skyblue', alpha=0.5)
plt.title("Area Plot of Customer Balances Over Time")
plt.xlabel("Index")
plt.ylabel("Balance")
plt.show()
```

```python
years = [2016, 2017, 2018, 2019, 2020]
balances = [20000, 25000, 30000, 35000, 40000]
balances1 = [22000, 27000, 32000, 37000, 42000]
balances2 = [21000, 26000, 31000, 36000, 41000]

plt.stackplot(years, balances, balances1, balances2, labels=['Branch A', 'Branch B', 'Branch C'], colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title("Stacked Area Plot of Balances Over Years")
plt.xlabel("Years")
plt.ylabel("Balances")
plt.legend(loc='upper left')
plt.show()
```
