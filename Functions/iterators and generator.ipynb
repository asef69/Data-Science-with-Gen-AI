```python
def square_number(n):
    result=[]
    for i in range(n):
        result.append(i*i)
    return result
```

```python
square_number(5)
```

```python
def square_number_generator(n):
    for i in range(n):
        yield i*i #creating generator using yield keyword
gen=square_number_generator(5)

print(next(gen))
print(next(gen))
print(next(gen))    
print(next(gen))
print(next(gen))
print(next(gen))  # This will raise StopIteration error
```

```python
def fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
fib_gen=fib(10)
for i in fib_gen:
    print(i)
```
